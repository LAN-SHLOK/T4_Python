## `super() method`
- If we write constructor in both parent and class class,the parent class container is not available in child class
- super() method is used to call parent class from the child class

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone class")
        self.price = price
        self.brand=brand
        self.camera=camera
    def buy(self):
        print("Buying a phone")
class Smartphone(Phone):
    def buy(self):
        print("Buying a smartphone")
        super().buy()
```

```python
s=Smartphone(20000,"I-Phone",13)
```

```python
s.buy()
```

```python
s.price
```

```python
s.brand
```

```python
s.camera
```

## Abstract class

- A class derived from ABC class which belongs to abc modules is known as abstract class in python
- ABC class is known as meta class which means that defines behaviour of other classes.
- Abstract class can have Concrete and abstract methods
- Abstract class needs to be extended and its methods implemented
- one cannot create object of an abstract class

```python
from abc import ABC,abstractmethod
class DefenceForce(ABC):
    def __init__(self):
        self.id=101
    @abstractmethod
    def area(self):
        pass
    def gun(self):
        print("The gun is AK - 47|")
```

```python
class Army(DefenceForce):
    def area(self):
        print("It will throw error",self.id)
        
```

```python
a=Army()
```

```python
a.area()
```

```python
a.gun()
```

```python
a.id
```

```python
class Navy(DefenceForce):
    def area(self):
        print("Area is sea",self.id)
```

```python
n=Navy()
```

```python
n.area()
```

```python
n.gun()
```
