## `super() method`
- If we write constructor in both parent and class class,the parent class container is not available in child class
- super() method is used to call parent class from the child class

````python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone class")
        self.price = price
        self.brand=brand
        self.camera=camera
    def buy(self):
        print("Buying a phone")
class Smartphone(Phone):
    def buy(self):
        print("Buying a smartphone")
        super().buy()
````

## Abstract class

- A class derived from ABC class which belongs to abc modules is known as abstract class in python
- ABC class is known as meta class which means that defines behaviour of other classes.
- Abstract class can have Concrete and abstract methods
- Abstract class needs to be extended and its methods implemented
- one cannot create object of an abstract class

````python
from abc import ABC,abstractmethod
class DefenceForce(ABC):
    def __init__(self):
        self.id=101
    @abstractmethod
    def area(self):
        pass
    def gun(self):
        print("The gun is AK - 47|")
````

````python
class Army(DefenceForce):
    def area(self):
        print("It will throw error",self.id)
        
````

## Method Resolution order in python
- It defines an order in which the class is searched in case of multiple inheritance
    - the C1 is the head of the class and the rest is tail
- It is based on C3 Linearization Algorithm is defines as MRO of a sum of any class C and the linear of parents and the list of parents
```
L(C(B1,B2,...BN)) = C + merge(L(B2)...BN))
```
- Here B1 to BN are the parent classes of C
1. Take the head of first list (B1) from left to right
2. If the head of selected class is not in the tail of the list,add it to the linear and remove it from the list
3. Repeat the process until the classes are removed or it is impossible to find good head

```
L(C)= C+ MERGE(L(D),L(F),DF)

```

```
L(B)= B + MERGE(L(D),L(E),DE)

```

```
L(A)= A + MERGE(L(B),L(C),BC)

= A + B + MERGE(DEO,CDFO,C)
= A + B + C +  MERGE(DEO,DFO)
= A + B + C + D + MERGE(EO,FO)

```

## Operator Overloading

- add=+
- sub=-
- mul=*
- truediv=/

````python
class Fraction:
    def __init__(self,x,y):
        self.num=x
        self.den=y
    def __add__(self,other):
        new_num=self.num*other.den+self.den*other.num
        new_den=self.den*other.den
        return f"{new_num}/{new_den}"
    def __sub__(self,other):
        new_num=self.num*other.den-self.den*other.num
        new_den=self.den*other.den
        return f"{new_num}/{new_den}"
    def __mul__(self,other):
        new_num=self.num*other.num
        new_den=self.den*other.den
        return f"{new_num}/{new_den}"
    def __truediv__(self,other):
        new_num=self.num*other.den
        new_den=self.den*other.num
        return f"{new_num}/{new_den}"
````

````python
f1=Fraction(1,3)
f2= Fraction(2,5)
print(f1.__add__(f2))
print(f1.__sub__(f2))
print(f1.__mul__(f2))
print(f1.__truediv__(f2))
````

## Write a program to create a class called Data having “value” as its data member. Overload the (>) and the (<) operator for the class. Instantiate the class and compare the objects using _lt_ and _gt_.

````python
class comparison:
    def __init__(self,x):
        self.num=x
    def __eq__(self,other):
        return self.num==other.num
    def __lt__(self,other):
        return self.num<other.num
    def __gt__(self,other):
        return self.num>other.num
````
